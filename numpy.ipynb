```python
import numpy as np
arr=np.array([1,2,3])
arr_2d=np.array([[1,2,3],[4,5,6]])
print(arr)
print(arr_2d)
```

```python
print(np.zeros((3,3)))
print()
print(np.ones((2,4)))
print()
print(np.eye(3))
print()
print(np.arange(0,10,2))
print()
print(np.linspace(0,1,5))
print()
print(np.random.rand(3,2))
print()
print(np.random.randn(3,2))

```

```python
arr=np.array([[1,2,3],[4,5,6]])
print(arr.shape)
print(arr.ndim)
print(arr.size)
print(arr.dtype)
```

```python
a=np.array([1,2,3])
b=np.array([4,5,6])
print(a+b)
print(a*b)
print(np.sin(a))
```

```python
a=np.array([[1,2],[3,4]])
b=np.array([[5,6],[7,8]])
print(np.dot(a,b))
print(a@b)
```

```python
arr=np.array([[1,2,3,4],[3,4,5,6]])
print(arr.sum())
print(arr.mean())
print(arr.max())
print(arr.min())
print(arr.std())
```

```python
arr=np.array([1,2,3,4,5])
print(arr[0])
print(arr[1:4])
print(arr[::2])
```

```python
arr=np.array([[1,2,3],[4,5,6]])
print(arr[0,1])
print(arr[:,1])
print(arr[1,:2])
```

```python
arr=np.array([1,2,3,4,5])
mask=arr>3
print(arr[mask])
```
